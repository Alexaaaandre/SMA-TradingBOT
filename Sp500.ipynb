```python
import pandas as pd
import os 
import numpy as np
import pandas_ta as ta
import plotly.graph_objects as go
```

```python
df = pd.read_csv(r"C:\Users\alrem\Desktop\données EURUSD\spy.csv")
```

```python
df
```

```python
df=df[df['Volume']!=0]
```

```python

df['SMA'] = ta.sma(df.Close, length=150)
```

```python
df.tail()
```

```python
df.tail(150)
```

```python
df=df[["Date","Open","Close","High","Low","SMA"]]
```

```python
df= df.tail(210)
```

```python
df
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
```

```python
df['SMA_30'] = df['Close'].rolling(window=30).mean()
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()
```

```python
fig.add_trace(go.Scatter(x=df['Date'], y=df['SMA_30'], mode='lines', name='SMA 30'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['SMA_50'], mode='lines', name='SMA 50'))
fig.add_trace(go.Scatter(x=df['Date'], y=df['SMA_200'], mode='lines', name='SMA 200'))
```
